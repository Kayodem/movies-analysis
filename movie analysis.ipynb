```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pylab as pl
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
df = pd.read_excel('C:\\Users\\Oladeji\\Downloads\\Movies.xlsx')
df
```

```python
df.drop(['Wikipedia URL', 'Director (2)', 'Cast (2)', 'Cast (3)', 'Cast (4)', 'Cast (5)', 'Genre (2)', 
         'Unnamed: 14'], inplace = True, axis = 1 )
```

```python
df.head()
```

```python
df.rename(columns={'Movie Title':'Title', 'Genre (1)':'Genre', 'Director (1)':'Director', 'Cast (1)':'Cast', 
                  'Budget ($)':'Budget','Box Office Revenue ($)':'Revenue'}, inplace=True)
```

```python
df.head()
```

```python
df['Profit'] = df['Revenue'] - df['Budget']
```

```python
df
```

```python
top_director = df['Director'].value_counts()
```

```python
top_director.head(20).reset_index()
```

```python
df.tail()
```

```python
david_profile = df[df['Director'] == 'David Ayer'].sort_values(by='Profit', ascending=False)
```

```python
david_profile
```

```python
high_revenue = df[['Title', 'Release Date', 'Genre', 'Director', 'Profit']].sort_values(by='Profit', ascending=False)
high_revenue
```

```python
df['Status'] = np.where(df.Revenue > df.Budget, 'gain','loss')
df
```

```python
genre_with_higher_revenue = df.groupby('Genre').max()[['Budget', 'Revenue', 'Profit']].sort_values(by='Profit', ascending=False).reset_index()
plt = sns.barplot(data=genre_with_higher_revenue.iloc[:10], y = 'Genre', x = 'Profit')
pl.title('genre with high profit')
pl.xlabel('profit')
pl.ylabel('genre')
pl.show()
```

```python
years_with_high_revenue = df.groupby('Director').max()[['Revenue']].sort_values(by='Revenue', ascending=False).reset_index()
plt = sns.barplot(data=years_with_high_revenue.iloc[:10], y = 'Director', x = 'Revenue',linewidth=2)
plt.show()
```

```python
sns.countplot(data=df, x = 'Status')
plt.show()
```

```python
sns.histplot(data=df, y = 'Release Date', hue='Status', element='step')
plt.show()
```
